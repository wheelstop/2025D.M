# 1) 함수 정의와 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 2 * x + 1

x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x, y, label = 'f(x) =2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Example of function graph')
plt.legend()
plt.grid(True)
plt.show() #우리 눈에 보여짐
```

# 2) 단사 함수 여부 확인

```python
def is_injective(f, domain):
  image = set()
  for x in domain:
    fx = f(x)
    if fx in image:
      return False
    image.add(fx)
  return True
#단사가 맞으면 True 아니면 False 값을 리턴함
domain = range(-10, 11)
print("Is it injective func?", is_injective(f, domain))
```

# 3)역함수 구하기 (단순한 경우)

```python
def inverse_f(y):
  return (y -1 ) / 2

y_values = np.linspace(-10, 10, 100)
x_values = inverse_f(y_values)

plt.plot(y_values, x_values, label='Inverse func f^{-1}(y) = (y - 1)/2', color='red')
plt.xlabel('y')
plt.ylabel('f^{-1}(y)')
plt.title('inverse func. graph')
plt.legend()
plt.grid(True)
plt.show()
```
